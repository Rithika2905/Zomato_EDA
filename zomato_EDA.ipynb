```python
for x in data.columns:
  if data[x].isnull().sum()>1:
    print(x, data[x].isnull().sum())
```

```python
data_cc = pd.read_excel('/content/Country-Code.xlsx')
data_cc.head()
```

```python
# Merge two dataset using merge keyword
# on keyword - using which field we need to combine dataset
# how - left or right join

final_data = pd.merge(data, data_cc, on='Country Code', how='left')
```

```python
final_data.Country.value_counts()
```

```python
country_name = final_data.Country.value_counts().index
country_name
```

```python
country_values = final_data.Country.value_counts().values
country_values
```

```python
# pie chart plott top 3 country using zomato
plt.pie(country_values[:3], labels=country_name[:3], autopct='%.2f%%')
plt.show()
```

```python
final_data.columns
```

```python
# size - Calculates the size of each group
# rest_index - Converts the grouped data back into a regular DataFrame by resetting the index.
# rename(columns={0:'count'}) - changes column with name 0 as count


final_data.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
final_data.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()
```

```python
rating = final_data.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'counts'})
rating
```

```python
rating.columns
```

```python
 # to increase graph size
matplotlib.rcParams['figure.figsize'] = (14,5)

sns.barplot(x='Aggregate rating', y='counts', data=rating)
```

```python
sns.barplot(x='Aggregate rating', y='counts', data=rating, hue='Rating color', palette=['White', 'Red','Orange','Yellow','Green','Green'])
```

```python
# sns.count plot - used for categorical variable

sns.countplot(x='Rating color', data=rating,hue='Rating color', palette=['White', 'Red','Orange','Yellow','Green','Green'])
```
